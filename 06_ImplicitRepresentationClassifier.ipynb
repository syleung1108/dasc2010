```python
import numpy as np
import sympy as smp
import matplotlib.pyplot as plt
from scipy.integrate import *
```

```python
def ExpFunctions(x,y):
    return np.exp(- x**2 - y**2) - np.exp(- (x-1)**2 - (y-1)**2)

x1 = np.arange(-2.0, 2.0, 0.1)
y1 = np.arange(-2.0, 2.0, 0.1)
X1, Y1 = np.meshgrid(x1, y1, indexing = 'ij')
Z1 = ExpFunctions(X1,Y1)

fig, ax = plt.subplots()
levels = np.arange(-1.2, 1.6, 0.2)
ax.contour(X1, Y1, Z1, levels)
plt.gca().set_aspect('equal', adjustable='box')
```

```python
fig, ax = plt.subplots()
ax.contour(X1, Y1, Z1, levels)
plt.gca().set_aspect('equal', adjustable='box')

x2 = np.arange(-2.0, 2.0, 0.25)
y2 = np.arange(-2.0, 2.0, 0.25)
X2, Y2 = np.meshgrid(x2, y2, indexing = 'ij')
Z2 = ExpFunctions(X2,Y2)
dZ = np.gradient(Z2,0.25,0.25)
plt.quiver(X2,Y2,dZ[0],dZ[1])
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
import numpy as np
import plotly.graph_objects as go
X, Y, Z = np.mgrid[-3:3:50j, -3:3:50j, -3:3:50j]
#X, Y, Z = np.meshgrid(np.arange(-3, 3, 50), np.arange(-3, 3, 50), np.arange(-3, 3, 50))
V = X**2+Y**2/4+Z**2/9
fig = go.Figure(data=go.Isosurface(
       x=X.flatten(),
       y=Y.flatten(),
       z=Z.flatten(),
       value=V.flatten(),
       opacity=0.5,
       isomin=1,
       isomax=1,
       surface_count=1,
       showscale=False, # remove colorbar
       caps=dict(x_show=False, y_show=False)
       ))
fig.show()
```
